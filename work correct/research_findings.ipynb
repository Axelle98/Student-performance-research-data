```python
import pandas as pd

# Load the data into a pandas DataFrame
file_path = '/content/drive/MyDrive/Questionnaires responses.xlsx'
df = pd.read_excel(file_path)
# Define age ranges
age_ranges = [(16, 20), (21, 24), (25, 30), (31, 35)]

# Categorize ages into ranges
def categorize_age(age):
    for lower, upper in age_ranges:
        if lower <= age <= upper:
            return f'{lower}-{upper}'
    return '100+'

df['Age Range'] = df['Age'].apply(categorize_age)


# Calculate frequencies of each age range
age_range_frequencies = df['Age Range'].value_counts().sort_index()

# Save frequencies to a new spreadsheet
output_file_path = 'age_range_frequencies.xlsx'
age_range_frequencies.to_excel(output_file_path)

import matplotlib.pyplot as plt

# Plot the histogram
plt.bar(age_range_frequencies.index, age_range_frequencies.values)
plt.xlabel('Age Range')
plt.ylabel('Frequency')
plt.title('Age Range Frequencies')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

Academic year frequencies


```python
import pandas as pd
file_path = '/content/drive/MyDrive/Questionnaires responses.xlsx'
df = pd.read_excel(file_path)

# Define academic year ranges
academic_year_ranges = [(1, 'Freshman'), (2, 'Sophomore'), (3, 'Junior'), (4, 'Senior')]

# Categorize academic years into ranges
def categorize_academic_year(year):
    for lower, label in academic_year_ranges:
        if year == lower:
            return label
    return 'Other'

df['Academic year Range'] = df['Academic year'].apply(categorize_academic_year)

# Calculate frequencies of each academic year range
academic_year_range_frequencies = df['Academic year Range'].value_counts().sort_index()

# Save frequencies to a new spreadsheet
output_file_path = 'academic_year_range_frequencies.xlsx'
academic_year_range_frequencies.to_excel(output_file_path)

import matplotlib.pyplot as plt

# Plot the histogram
plt.bar(academic_year_range_frequencies.index, academic_year_range_frequencies.values)
plt.xlabel('Academic year Range')
plt.ylabel('Frequency')
plt.title('Academic year Range Frequencies')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

time spent on academic activities

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Load the dataset from Excel
file_path = '/content/Questionnaires responses.xlsx'
df = pd.read_excel(file_path)

# Extract the column for analysis
data_column = df['time spent academic activities daily']

# Perform the Shapiro-Wilk test for normality
statistic, p_value = stats.shapiro(data_column)

# Set significance level
alpha = 0.05

# Print the results
print(f"Shapiro-Wilk Test - Statistic: {statistic}, p-value: {p_value}")
if p_value < alpha:
    print("The data is not normally distributed (reject null hypothesis)")
else:
    print("The data is normally distributed (fail to reject null hypothesis)")

# Create a histogram to visualize the distribution
plt.hist(data_column, bins=10, edgecolor='black')
plt.xlabel('Time Spent on Academic Activities Daily')
plt.ylabel('Frequency')
plt.title('Histogram of Time Spent on Academic Activities')
plt.show()
```

types of social media : mean , standard deviation, sigma, t-scores

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the data into a pandas DataFrame
file_path = '/content/Questionnaires responses.xlsx'
df = pd.read_excel(file_path)

# Column name for social media platforms
platforms_column = 'types of social media'

# Calculate frequencies of each social media platform
platforms_frequencies = df[platforms_column].apply(lambda x: x.split(', ')).explode().value_counts()

# Calculate the total number of users
total_users = platforms_frequencies.sum()

# Calculate the mean and standard deviation of users per platform
mean_users = platforms_frequencies.mean()
std_users = platforms_frequencies.std()

# Calculate the mean and standard deviation of the frequencies
mean_frequency = platforms_frequencies.mean()
std_deviation = platforms_frequencies.std()

# Calculate the sample standard deviation (sigma)
sigma = np.sqrt(((platforms_frequencies - mean_frequency) ** 2).sum() / (platforms_frequencies.size - 1))

# Choose a specific value to calculate the t-score for (e.g., value = 100)
value = 180
# Calculate the t-score
t_score = (value - mean_frequency) / (sigma / np.sqrt(platforms_frequencies.size))

# Print the mean and standard deviation of frequencies
print(f"Mean Frequency: {mean_frequency:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")

# Print the number of users for each platform
print("\nNumber of Users per Social Media Platform:")
print(platforms_frequencies)

# Print the total number of users
print(f"\nTotal Number of Users: {total_users}")

# Print the mean and standard deviation of users
print(f"Mean Users per Platform: {mean_users:.2f}")
print(f"Standard Deviation of Users: {std_users:.2f}")

# Print the calculated sample standard deviation (sigma) and t-score
print(f"\nSample Standard Deviation (Sigma): {sigma:.2f}")
print(f"T-Score for Value {value}: {t_score:.2f}")
```

correlation between academic performance and social media types


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Load the data into a pandas DataFrame
file_path = '/content/responses_1.sav'
df = pd.read_spss(file_path)

# Column names
platforms_column = 'typesofsocialmedia'
academic_performance_column = 'academicperformance'


# Calculate frequencies of each social media platform
platforms_frequencies = df[platforms_column].apply(lambda x: x.split(', ')).explode().value_counts()

# Loop through each social media type
for social_media_type in platforms_frequencies.index:
    # Select rows where the current social media type is present
    selected_rows = df[df[platforms_column].str.contains(social_media_type, na=False)]

    # Calculate the mean frequency for the current social media type
    mean_frequency = platforms_frequencies[social_media_type]

    # Calculate the sample standard deviation (sigma)
    sigma = np.sqrt(((selected_rows[academic_performance_column] - mean_frequency) ** 2).sum() / (selected_rows[academic_performance_column].size - 1))

    # Calculate the t-score for a specific value (e.g., value = 3)
    t_score = (mean_frequency - 3) / (sigma / np.sqrt(selected_rows[academic_performance_column].size))

    print(f"Social Media Type: {social_media_type}")
    print(f"Sigma: {sigma:.2f}")
    print(f"t-score for value 3: {t_score:.2f}")
    print()
```

```python
import pyreadstat
import pandas as pd
import pingouin as pg
from graphviz import Digraph


# Read the SPSS dataset
data, metadata = pyreadstat.read_sav('/content/responses_1.sav')

# Convert to a pandas DataFrame
df = pd.DataFrame(data)

# Calculate covariance and correlation matrices
cov_matrix = data[['Academicyear', 'Age']].cov()
corr_matrix = data[['Academicyear', 'Age']].corr()

# Draw the model (a simple example)
dot = Digraph(comment='Model')
dot.node('A', 'Availability of Digital Resources')
dot.node('B', 'Time Spent on Social Media')
dot.node('C', 'Time Spent on Academic Activities')
dot.node('D', 'Self-Efficacy')
dot.node('E', 'Current GPA')
dot.node('F', 'Academic Performance')
dot.edges(['AB', 'AC', 'AD', 'AE', 'BC', 'BF', 'CE', 'CF', 'DE', 'DF'])

# Calculate Chi-Square, Degrees of Freedom, and Probability
chi2, p_value, dof, _ = pg.chi2_independence(data[['Academicyear', 'Age']])

# Print the results
print(f'Chi-Square: {chi2}')
print(f'Degrees of Freedom: {p_value.shape[0] - 1}')
print(f'Probability: {p_value}')

# Display the model
dot.render('model', format='png', view=True)
```
